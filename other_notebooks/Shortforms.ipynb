```python
from adeft.modeling.find_anomalies import AdeftAnomalyDetector
```

```python
task_stmts = list(channel_texts.values())[2]['task']
task_stmts = [x[0] for x in task_stmts]
```

```python
_, task_content = get_text_content_from_stmt_ids(task_stmts)
```

```python
_, kcnk3_content = get_text_content_from_pmids(pmids['KCNK3'])
```

```python
anomaly_detector = AdeftAnomalyDetector(blacklist=['task', 'task1', 'kcnk3'])
```

```python
param_grid = {'nu': [0.1, 0.2, 0.4], 'max_features': [100], 'ngram_range': [(1, 1)]}
```

```python
list(kcnk3_content.values())
```

```python
anomaly_detector.cv(list(kcnk3_content.values()), [], param_grid, cv=5, n_jobs=8)
```

```python
anomaly_detector.specificity
```

```python
preds = anomaly_detector.predict(task_content.values())
```

```python
sum(preds)/len(preds)
```

```python
anomaly_detector.confidence_interval(task_content.values(), sensitivity=0.75)
```
